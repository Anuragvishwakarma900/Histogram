```python
data = [np.random.normal(0,std,100) for std in range(0,5)]
data
```

```python
lables=['pythons','c++','java','Mysql']
sizes = [300,420,250,290]
colors=['cyan','black','magenta','lightgreen']
explod=(0.1,0.2,0.0,0.1)
plt.pie(sizes,explode=explod,labels=lables,colors=colors, autopct ='%1.1f%%', shadow=0.1)
plt.show()
```
